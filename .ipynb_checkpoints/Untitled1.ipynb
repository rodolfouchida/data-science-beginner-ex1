```python
from itertools import permutations
from sklearn.cluster import KMeans
import numpy as np
import sklearn.decomposition as dec
import sklearn.cluster as clu
import sklearn.datasets as ds
import sklearn.model_selection as ms
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X, y = ds.make_blobs(n_samples = 500, centers =5, cluster_std = 1.2)
plt.scatter(X[:,0], X[:,1], s = 10)
```

```python
kmeans = KMeans(n_clusters = 5
                , random_state = 0).fit(X)
plt.scatter(X[:,0], X[:,1], c = kmeans.labels_, s = 10)
plt.title('Resultados K-Means')
plt.show()
```

```python
Nc = range(1,20)
kmeans = [KMeans(n_clusters = i) for i in Nc]
score = [kmeans[i].fit(X).score(X) for i in range(len(kmeans))]
plt.plot(Nc, score, marker = 'o')
plt.xlabel('Numeros de clusters')
plt.title('Elbow Curve')
plt.show()
```

```python
from sklearn.cluster import SpectralClustering

af = SpectralClustering(n_clusters = 5, eigen_solver='arpack', affinity="nearest_neighbors").fit(X)

plt.scatter(X[:, 0], X[:, 1], c=af.labels_, s=10)
plt.title('Resultado Spectral Cluster')

plt.show()
```

```python
X, y = ds.make_circles(n_samples = 800, factor = .4, noise = .09)
plt.scatter(X[:,0],X[:,1], s = 8)
```

```python
kmeans = KMeans(n_clusters=2, random_state=0).fit(X)

plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, s=10)
plt.title('Resultado K-Means')
plt.show()
```

```python
af = SpectralClustering(n_clusters = 2, eigen_solver='arpack', affinity="nearest_neighbors").fit(X)

plt.scatter(X[:, 0], X[:, 1], c=af.labels_, s=10)
plt.title('Resultado Spectral Clustering')

plt.show()
```
