```python
from itertools import permutations
from sklearn.cluster import KMeans
import numpy as np
import sklearn
import sklearn.decomposition as dec
import sklearn.cluster as clu
import sklearn.datasets as ds
import sklearn.model_selection as ms
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Gerando dados para a clusterização

X, y = ds.make_blobs(n_samples=500,
                     centers=3,
                     cluster_std=1.5,
                     )

plt.scatter(X[:, 0], X[:, 1], s=10);
```

```python
# Aplicando o algortimo de clusterização Kmeans

kmeans = KMeans(n_clusters=3, random_state=0).fit(X)

plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, s=10)
plt.title('Resultado K-Means')
plt.show()
```

```python
# Validando o modelo K-means

Nc = range(1, 20)

kmeans = [KMeans(n_clusters=i) for i in Nc]

score = [kmeans[i].fit(X).score(X) for i in range(len(kmeans))]

plt.plot(Nc,score, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Score')
plt.title('Elbow Curve')
plt.show()
```

```python
# Testando o Spectral

from sklearn.cluster import SpectralClustering

af = SpectralClustering(n_clusters = 2, eigen_solver='arpack', affinity="nearest_neighbors").fit(X)

plt.scatter(X[:, 0], X[:, 1], c=af.labels_, s=10)
plt.title('Resultado Spectral Cluster')

plt.show()
```

```python
# Complicando o shape dos clusters

X, y = ds.make_circles(n_samples=800, factor=.1,
                                      noise=.08)
plt.scatter(X[:, 0], X[:, 1], s=8);
```

```python
# Como o k-means performa

kmeans = KMeans(n_clusters=2, random_state=0).fit(X)

plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, s=10)
plt.title('Resultado K-Means')
plt.show()
```

```python
# Vamos testar o Spectral

af = SpectralClustering(n_clusters = 2, eigen_solver='arpack', affinity="nearest_neighbors").fit(X)

plt.scatter(X[:, 0], X[:, 1], c=af.labels_, s=10)
plt.title('Resultado Spectral Clustering')

plt.show()
```
